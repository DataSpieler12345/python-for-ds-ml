# Problem Statement

Load the load_diabetes datasets internally from sklearn and check for any missing value or outlier data in the 'data' column. if any irregularities found, treat them acordingly.

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
dataset
```

```python
dataset.data
```

```python
dataset.target #target or label for training data
```

```python
#extract the column name of the data set
dataset['feature_names']
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(data=np.c_[dataset['data'],dataset['target']], columns = dataset['feature_names']+['target'])
df
```

```python
# Important to check if there any missing value
df.isnull().any
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

for column in df:
    plt.figure()
    df.boxplot([column])
```
