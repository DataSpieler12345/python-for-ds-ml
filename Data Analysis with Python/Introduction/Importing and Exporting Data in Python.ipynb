**To read any data using Pythons pandas** package, there two important factor to consider: 

**format** = is the way data is encoded. We can usually tell different encoding schemes by looking at the ending of the file name. Some common encodings are csv, json, xlsx, hdf and so forth.

and 

**file path** = tell us where the data is stored. Usually its stored either on the computer we are using, or online on the internet.
