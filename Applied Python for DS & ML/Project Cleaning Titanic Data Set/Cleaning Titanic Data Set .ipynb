#### Identifying Data Cleaning Issues 

(1) There are NaN values in Age column.

(2) We need not require columns like PassengerId, Name, Cabin and Ticket 

(3) Apart from Survived Column having values '0' and '1', creating a new column with Class/Categorical Status of Survival is needed mentioning 'Survived' or  'Did Not Survived'.

(4) Fare value i.e 512.32 in the data is an outlier

```python
print('Mean Fare is', dp['Fare'].mean())
print('Minimum Fare is', dp['Fare'].min())
print('Maximum Fare is', dp['Fare'].max())
```

```python
# let's see that how many unique values are there in our dataset.
dp['Fare'].unique()
```

```python
#values > 300
dp[dp['Fare'] > 300]
```

(2) We need not require columns like PassengerId, Name, Cabin and Ticket 

```python
dp = dp.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis =1)
```

```python
dp.head()
```

(1) There are NaN values in Age column.

```python
#first data cleaning issue: Dealing with the NaN values by removing them
dp = dp.dropna()
```

```python
# We check our first cleaning issue is resolved or not.
# And it is resolve. Now we do not have any NaN values in our data. 
dp.info()
```

(3) Apart from Survived Column having values '0' and '1', creating a new column with Class/Categorical Status of Survival is needed mentioning 'Survived' or  'Did Not Survived'.

```python
# Creating a Column with Class/Categorical Status of Survival instead of numbers 
dp['Survival_Status'] = dp['Survived'].replace({1: 'Survived', 0: 'Dit not Survived'})
```

```python
dp.head()
```

(4) Fare value i.e 512.32 in the data is an outlier

```python
dp = dp[dp['Fare'] < 512]
```

```python
#there no row with values > 300
dp[dp['Fare'] > 300]
```
