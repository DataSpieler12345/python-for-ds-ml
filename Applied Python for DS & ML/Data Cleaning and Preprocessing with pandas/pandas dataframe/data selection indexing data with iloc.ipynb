#### Import pandas

```python
import pandas as pd
```

#### Load the data

```python
data = pd.read_csv('./data/Lending-company.csv', index_col = 'LoanID')
lending_co_data = data.copy()
lending_co_data.head()
```

```python
lending_co_data[1] # KeyError[1]
```

```python
lending_co_data[0,1] # KeyError(0, 1)
```

```python
lending_co_data['Product']
```

#### iloc[]

```python
lending_co_data.iloc[1]
```

```python
#indexing starts from 0!
lending_co_data.iloc[1, 3]
```

```python
# .iloc[] has been programmed to deliver the desired portion of the dataset whether we provide one or two lcoation specifiers
lending_co_data.iloc[1,:]
```

```python
lending_co_data.iloc[:, 3]
```

```python
type(lending_co_data.iloc[1, 3])
```

```python
type(lending_co_data.iloc[1, :])
```

```python
type(lending_co_data.iloc[:, 3])
```

```python
lending_co_data.iloc[[1, 3], :]
```

```python
lending_co_data.iloc[:, [1, 3]]
```
