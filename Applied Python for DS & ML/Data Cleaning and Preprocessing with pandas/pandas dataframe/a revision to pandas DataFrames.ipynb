#### A revision to pandas DataFrames 

#### Import Libraries

```python
import pandas as pd
import numpy as np
```

#### Create an Array

```python
array_a = np.array([[3, 2, 1], [6, 3, 2]])
array_a
```

```python
array_a.dtype
```

#### check if 1D or 2D arrays?

```python
# Verificar si es 1D o 2D
if array_a.ndim == 1:
    print("unidimensional (1D)")
elif array_a.ndim == 2:
    print("bidimensional (2D)")
else:
    print("more than 2 dimensions")
```

#### created array_a saved as pandas DF

```python
pd.DataFrame(array_a)
```

```python
type(pd.DataFrame(array_a))
```

#### given a columns

```python
df = pd.DataFrame(array_a, columns = ['Column 1', 'Column 2', 'column 3'])
df
```

```python
df = pd.DataFrame(array_a, columns = ['Column 1', 'Column 2', 'column 3'], index = ['Row 1', 'Row 2'])
df
```
